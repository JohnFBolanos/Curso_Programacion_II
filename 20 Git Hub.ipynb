Cuando el entorno est√© activo, la terminal mostrar√° algo como:
```bash
(.venv)
C:\Users\John\proyecto> (main)
```

üí° Todo lo que instales con pip quedar√° dentro de la carpeta `.venv` , la cual no se debe subir a GitHub (ya estar√° incluida en el `.gitignore`).

## üß© Qu√© puedes hacer ahora

Una vez activado el entorno virtual (cuando veas `(.venv)` al inicio de tu l√≠nea en la terminal), puedes verificar y preparar tu entorno de trabajo siguiendo estos pasos:

---

###  Verifica la versi√≥n de Python dentro del entorno

```bash
python --version
```

###  Instala una librer√≠a de prueba (ejemplo: NumPy)

###  Comprueba que est√° instalada solo en el entorno

üí° Si aparece numpy en la lista, significa que la instalaci√≥n fue exitosa y est√° contenida dentro del entorno.

## 4Ô∏è‚É£ Instalar las dependencias del proyecto

Si el proyecto tiene un archivo requirements.txt, inst√°lalas con:

```bash
pip install -r requirements.txt
```

En nuestro caso nuestro proyecto no tiene ese archivo por lo tanto lo creamos y guardarmos la primera libreria numpy:


```bash
pip freeze > requirements.txt
```

## üß© Flujo de trabajo con el archivo `requirements.txt`

Cuando creas o actualizas el archivo `requirements.txt` dentro del entorno virtual, Git lo detecta como un **archivo nuevo o modificado**.  
Esto es normal: el archivo es parte importante del proyecto y **s√≠ debe subirse al repositorio** (a diferencia de la carpeta `.venv/`, que nunca se sube).

---

### 1 Verifica el estado del repositorio

Despu√©s de crear o actualizar el archivo `requirements.txt`, ejecuta:

```bash
git status
```

Ver√°s algo como esto:



image.png

üí° El color rojo significa que Git detecta el archivo, pero a√∫n no est√° agregado al √°rea de seguimiento (staging area).

### 2 Agrega el archivo al control de versiones


```bash
git add requirements.txt
```
o puede usar add . para agregar TODOS los archivos pendientes:

```bash
git add .
```
Luego puedes verificar de nuevo:

```bash
git status
```

Ahora debe verse en **verde**, indicando que est√° listo para confirmar (commit):

### 3 Realiza el commit


```bash
git commit -m "Agrego archivo requirements.txt con dependencias del entorno"
```

### 4 Sube los cambios a GitHub


```bash
git push
```
üí° Ahora el archivo requirements.txt estar√° disponible para todo el equipo en GitHub.

---

### Uso colaborativo del archivo requirements.txt


```bash
pip install -r requirements.txt

```
Esto garantiza que todos los miembros trabajen con el mismo entorno y las mismas versiones de las dependencias.





### üí° Ejemplo completo del flujo

#### Activar entorno virtual

```bash
source .venv/Scripts/activate

```
#### Instalar o actualizar librer√≠as

```bash
pip install numpy pandas matplotlib

```

```bash
git commit -m "Actualizo dependencias en requirements.txt"

```

#### Subir al repositorio remoto

```bash
git push
```


### üß† Consejo

En proyectos colaborativos, se recomienda:

* Actualizar requirements.txt solo cuando cambien librer√≠as.

* Avisar al equipo para que todos ejecuten luego:


```bash
git pull
pip install -r requirements.txt
```

Esto garantiza que todos los entornos virtuales est√©n sincronizados y nadie tenga errores por librer√≠as faltantes o versiones distintas.
