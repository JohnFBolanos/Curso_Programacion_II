# ITERADORES

Un iterador es un objeto en Python que permite recorrer una colección de elementos uno a uno, sin la necesidad de cargar todos los elementos en la memoria al mismo tiempo, pero solo se pueden usar UNA SOLA VEZ.

* iter()
* next()
* reversed()
* enumerate()
* zip()
* filter()
* map()


## iter()

Este método devuelve el propio objeto iterador y se llama una vez al inicio de un bucle para inicializar el iterador

```python
A = iter(range(10))
B = list(range(10))
print(type(A),type(B))
#Tanto A como B tiene los mismos elementos use un for para ver esto...
```

## next()

Este método se utiliza para obtener el siguiente elemento de la secuencia. Cuando no hay más elementos que recorrer, se debe levantar una excepción

```python
#Correr varias veces next(A) para ver como se accede a los valores del iterador

next(A)

for k in A:
    print(k)
```

## filter()

La función `filter()` crea un iterador que genera los elementos de la secuencia original que cumplen con el criterio especificado por la función. Los elementos que no cumplen con el criterio son omitidos.

```python
numeros=[x for x in range(2,15)]
comprueba_par=lambda x:x%2==0
numeros_pares = filter(comprueba_par, numeros)
print(type(numeros_pares))
print(next(numeros_pares))
print(next(numeros_pares))
NuevaLista=list(numeros_pares)
NuevaLista
```

## enumerate

La función `enumerate()` en Python es una función incorporada que se utiliza para recorrer una secuencia (como una lista, una tupla o una cadena) al mismo tiempo que realiza un seguimiento del índice (la posición) de cada elemento en la secuencia. Esto es especialmente útil cuando necesitas acceder tanto al valor como al índice de los elementos durante un bucle.

La función `enumerate()` toma una secuencia como argumento y devuelve un iterador que genera **tuplas**, donde cada tupla contiene dos elementos: el índice (empezando desde 0 por defecto) y el valor del elemento correspondiente en la secuencia.

```python
Nombres=["John", "Juana", "Carlos", "Ana", "Lola"]

numerados=list(enumerate(Nombres))
numerados=enumerate(Nombres)
print(numerados)


for elemento in numerados:
    print(elemento)

# si dejamos numerados como un iterador y no como lista el siguiente for no tiene nada para mostrar ya que uso todos los elementos del iterador.

for k , i in numerados:
    print(k,"es", i)

```

```python
### enumerate(iterador, indice)
Nombres=["John", "Juana", "carlos", "Ana", "Lola"]
numerados2=enumerate(Nombres,5)

#for num in numerados2:
 #   print(num)
next(numerados2), next(numerados2)
```

```python
##Listas por comprensión
lista1=list(enumerate (Nombres, 3))
print(lista1)

lista2=[[x,y] for x , y in enumerate (Nombres, 3)]
print(lista2)

lista3=[list(k) for k in enumerate(Nombres,3)]
print(lista3)
```

```python
# se puede convertir enumerate en un diccionario
diccionario=dict(enumerate(Nombres,4))
diccionario
```

```python
# enumerate con range

for k in enumerate(range(8,30,5),4):
    print(k)
```

```python
# enumerate con cadenas

for k, j in enumerate("Viernes Loco",4):
    print(k, j)
```

### Ejercicio 2

Cree una función que recibe una lista de números naturales y retorne una lista con los **índices** de todas las ocurrencias del número mayor de la lista (use la función `enumerate()` )

```python
### Solución ejercicio 2
def fun(lista):
    rta=[]
    M=max(lista)
    for pos , num in enumerate(lista,10) :
        if num == M:
            rta.append(pos)
    return rta
L=[2,5,1,5,3,5,5]
fun(L)
```

# GENERADORES

Son funciones simples que generan iteradores

```python
def generador(limite):
    contador = 2
    while contador < limite:
        yield contador
        contador = contador + 1

A = generador(5)

type(A)
next(A),next(A)
#b = list(A)
#b

```

```python
next(A)
```

```python
def mi_enumerate(iterable, inicio=0):
    contador = inicio
    for elemento in iterable:
        yield contador, elemento
        contador = contador + 1

nombres = ["Juan","Luisa", "Carlos", "Pedro"]

A = mi_enumerate(nombres,10)
print(next(A),next(A))
B=list(A)
B
```

### Ejercicio 5: Generador de números pares filtrados

Diseña un **generador en Python** que produzca los **n primeros números pares mayores que 10**,  asegurándote de que cada número sea **par** y **no sea múltiplo de 3**.

```python
### Solución ejercicio 5
def generador(n):
    for i in range(n):
        if i %2 == 0 and i%3 != 0 and i > 10 :
            yield i

a = generador(20)
next(a)
```

```python
def generador(num):
    cantidad=0
    inicio=11

    while cantidad < num:
        if inicio%2==0 and inicio%3!=0:
            yield inicio
            cantidad=cantidad+1
        inicio=inicio+1

M=generador(5)
```

```python
next(M)
```
