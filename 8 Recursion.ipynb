## Para entender la recursividad debemos entender las pilas

Una **pila** es una de las estructuras de datos más simples en informática.  
Almacena múltiples valores, como lo hace una lista, pero con una diferencia fundamental:  
las pilas limitan la forma en que se agregan o eliminan elementos, permitiendo hacerlo *SOLO* desde la parte superior de la pila.

- **Agregar valores** → se denomina **insertar** valores en la pila.  
- **Eliminar valores** → se denomina **extraer** valores de la pila.

Una función recursiva es una función que se llama a sí misma.

## Caso Base y Recursivo

Toda función recursiva debe tener un caso base y al menos un caso recursivo, con el caso base nos permite que no haya un desbordamiento de pila y si no hay un caso recursivo nunca se llamara a si misma.

```python
def fun(variable):
  print(f'fun({variable}) evaluada.')
  if not variable:
    #CASO BASE
    print('Retorna caso base.')
    return
  else:
  # CASO RECURSIVO
    #fun(False)
    print('retorna caso recursivo.')
    return fun(False)

#print('Evaluando fun(False):')
#fun(False)
print('Evaluando fun(True):')
fun(True)
```

```python
def conteo(number):
  print(number)
  if number == 0:
    # CASO BASE
    print('Llegó al caso base.')
    return
  else:
  # CASO RECURSIVO
    #conteo(number - 1)
    print(number, 'retornado')
    return conteo(number - 1)
conteo(3)
```

```python
def sumar(n):
  if n <= 3:
    return n
  else:
    return n + sumar(n-1)

print(sumar(6))
```

```python
def f(a,b):
    if a < 1:
        return 3
    elif b <= 0:
        return -2
    elif a%2 == 0:
        return a + f(b-1,a-1)
    else:
        return b + f(a-1,b-1)

print(f(3,5))









```

```python
def fact1(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact1(n-1)

def fact2(n):
    rta=1
    for k in range(1,n+1):
        rta=rta*k
    return rta

fact2(5)




```
