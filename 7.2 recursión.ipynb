```python


def fun1(n):
  if n <= 4:
    return n
  else:
    return n - fun1(n-3)

print(fun1(12))
```

```python
def fun2(n):
  if n <= 3:
    return n
  else:
    return fun2(n-1) - fun2(n-3)

print(fun2(7))
```

```python
def fun3(a,b):
  a , b = b % a , a
  if a == 0:
    return b
  else:
    return fun3(a,b)

print(fun3(60,24))
```

```python
def fun4(str):
  if len(str) < 2 :
    return ""
  else:
    return str[0] + fun4(str[1:]) + str[0]

print(fun4("Hola"))
```

```python
def fun5(n):
  if n == 0 :
    return 5
  if n == 1 :
    return 2
  else: return n-fun5(n-2)

print(fun5(6))
```

```python
def fun6(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        return max(lst[0], fun6(lst[1:]) )

print(fun6([4,5,2,-4]))
```

```python
def fun7(n):
  if n==2:
    return n
  if n > 0:
    print(2*n)
    return fun7(n-1)

print(fun7(5))
```

```python
def fun8(a,b,c):
    if a*b*c == 0:
        return a+b-c
    else:
        return c-b+a*fun8(c-1,a-1,b-1)

print(fun8(3,4,5))
```

```python
def fun9(n,t):
  if t == 0:
    return 3
  elif n <= 1:
    return 2
  else:
    return n - t + fun9(n-2,t-1)

print( fun9(6,4))
```

```python
def fun10(n):
  if n == 0 :
    return 3
  elif n == 1:
    return 1
  else:
    return 3*fun10(n-1)-fun10(n-2)

print(fun10(4))
```
