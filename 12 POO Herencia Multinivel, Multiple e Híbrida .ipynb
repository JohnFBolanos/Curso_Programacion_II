### Herencia Multinivel:

image.png

La herencia multinivel es un concepto en el que una clase puede heredar de otra clase que, a su vez, hereda de otra clase. Es una extensi√≥n de la herencia simple. En este caso, las clases forman una jerarqu√≠a en niveles, donde cada nivel representa una nueva capa de especializaci√≥n o abstracci√≥n.



```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        print(f"{self.nombre} hace un sonido jajaja")

class Ave(Animal):

    def volar(self):
        print(f"{self.nombre} est√° volando")

class Pato(Ave):

    def hacer_sonido(self):
        print("Cua")

class PatoDomestico(Pato):

    def ser_mascota(self):
        print(f"{self.nombre} es una mascota")


# Crear una instancia de PatoDomestico y utilizar m√©todos
mi_pato_domestico = PatoDomestico("Donald")
mi_pato_domestico.hacer_sonido()
#mi_pato_domestico.volar()
#mi_pato_domestico.ser_mascota()
#Animal.hacer_sonido(mi_pato_domestico)
```


**Ventajas de la herencia multinivel:**
1. **Estructura Jer√°rquica Clara:** Facilita la organizaci√≥n de las clases en una jerarqu√≠a m√°s f√°cil de entender.

2. **Especializaci√≥n Gradual:** Permite una especializaci√≥n gradual de clases a medida que se desciende en la jerarqu√≠a.





## Herencia M√∫ltiple:

image.png

La herencia m√∫ltiple es un concepto en el que una clase puede heredar atributos y m√©todos de m√°s de una clase base. En un lenguaje de programaci√≥n que admite herencia m√∫ltiple, una clase derivada puede tener varias clases base. Esto permite la reutilizaci√≥n de c√≥digo y la construcci√≥n de jerarqu√≠as de clases m√°s complejas.


```python
# Ejemplo HERENCIA MULTIPLE

class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        print(f"{self.nombre} hace un sonido")

class Volador:
    def volar(self):
        print(f"{self.nombre} est√° volando")

class Nadador:
    def nadar(self):
        print(f"{self.nombre} est√° nadando")

class Pato(Animal, Volador, Nadador):

    def info(self):
        print(f"Soy un pato")

    def hacer_sonido(self):
        print("Cua")

# Crear una instancia de Pato y utilizar m√©todos
mi_pato = Pato("Donald")
mi_pato.hacer_sonido()
#mi_pato.volar()
#mi_pato.nadar()
#mi_pato.info()
#Animal.hacer_sonido(mi_pato)
```


**Ventajas de la herencia m√∫ltiple:**

1. **Reutilizaci√≥n de C√≥digo:** Permite que una clase herede funcionalidades de varias fuentes, promoviendo la reutilizaci√≥n de c√≥digo.
  
2. **Combinaci√≥n de Caracter√≠sticas:** Permite combinar caracter√≠sticas de diferentes clases en una nueva clase.

**Desaf√≠os y consideraciones:**

1. **Ambig√ºedades:** Pueden surgir ambig√ºedades si dos o m√°s clases base tienen atributos o m√©todos con el mismo nombre.

2. **Complejidad:** A medida que se aumenta la cantidad de clases base, la jerarqu√≠a puede volverse compleja y dif√≠cil de entender y mantener.

## Diferencias:

1. **N√∫mero de Clases Base:**
   - En la herencia m√∫ltiple, una clase puede heredar de varias clases base.
   - En la herencia multinivel, una clase hereda de una clase base directamente y puede haber m√∫ltiples niveles de herencia.

2. **Organizaci√≥n de la Jerarqu√≠a:**
   - La herencia m√∫ltiple permite una organizaci√≥n m√°s horizontal de la jerarqu√≠a.
   - La herencia multinivel implica una organizaci√≥n m√°s vertical de la jerarqu√≠a.

En resumen, ambas herencias tienen sus ventajas y desaf√≠os. La elecci√≥n entre herencia m√∫ltiple y multinivel depende del dise√±o espec√≠fico del sistema y de los requisitos de flexibilidad y organizaci√≥n de clases.


## Herencia H√≠brida:


image.png

```python
#Ejemplo 1:

class A:
    def hablar(self):
        print("Hola desde A")
    pass

class B(A):
    def hablar(self):
        print("Hola desde B")
    pass

class F:
    def hablar(self):
        print("Hola desde F")
    pass

class C(F):
    def hablar(self):
        print("Hola desde C")
    pass

class D(B,C):
    def hablar(self):
        print("Hola desde D")
    pass

objeto=D()
objeto.hablar()
#D.mro()
```



La funci√≥n `mro()` en Python (Method Resolution Order) se usa en programaci√≥n orientada a objetos para mostrar el orden en el que se buscan los m√©todos y atributos en una jerarqu√≠a de clases cuando una clase hereda de varias clases (herencia m√∫ltiple o h√≠brida).

La funci√≥n mro() devuelve una lista de clases en el orden en que Python las buscar√° para resolver un m√©todo o atributo.


```python
#Ejemplo 1:

class A:
    def hablar(self):
        print("Hola desde A")
    pass

class B(A):
    def hablar(self):
        print("Hola desde B")
    pass

class F:
    def hablar(self):
        print("Hola desde F")
    pass

class C(F):
    def hablar(self):
        print("Hola desde C")
    pass

class D(C,B):
    def hablar(self):
        print("Hola desde D")
    pass


objeto=D()

objeto.hablar()
#print(D.mro())

#A.hablar(objeto) # quiero que el objeto hable desde B
```

En Python, la sintaxis para los atributos en herencia m√∫ltiple es bastante sencilla. Cuando una clase hereda de m√∫ltiples clases base, los atributos de cada clase base est√°n disponibles en la clase derivada. Aqu√≠ hay un ejemplo:


```python
class ClaseBase1:
    def __init__(self, atributo_base1):
        self.atributo_base1 = atributo_base1

class ClaseBase2:
    def __init__(self, atributo_base2):
        self.atributo_base2 = atributo_base2

class ClaseDerivada(ClaseBase1, ClaseBase2):
    def __init__(self, atributo_base1, atributo_base2, atributo_derivado):
        # Llamadas a los constructores de las clases base
        ClaseBase1.__init__(self, atributo_base1)
        ClaseBase2.__init__(self, atributo_base2)
        # Ojo cuando solo se hereda una clase se usa super() y no se usa self

        # Atributo espec√≠fico de la clase derivada
        self.atributo_derivado = atributo_derivado

# Crear una instancia de la clase derivada
objeto_derivado = ClaseDerivada("Valor1", "Valor2", "ValorDerivado")

# Acceder a los atributos
print(objeto_derivado.atributo_base1)
print(objeto_derivado.atributo_base2)
print(objeto_derivado.atributo_derivado)
```

```python
# ===============================
# Taller de Herencia en Python
# Zool√≥gico Digital
# ===============================

# Clase base
class Animal:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def comer(self):
        print(f"{self.nombre} est√° comiendo üêæ")

    def envejecer(self):
        self.edad += 1
        print(f"{self.nombre} ahora tiene {self.edad} a√±os")

# Subclase: Carn√≠voro (herencia simple)
class Carnivoro(Animal):
    def __init__(self, nombre, edad, tipo_diente):
        super().__init__(nombre, edad)
        self.tipo_diente = tipo_diente

    def cazar(self):
        print(f"{self.nombre} est√° cazando con sus dientes {self.tipo_diente}")

# Subclase: Felino (herencia multinivel: Animal -> Carn√≠voro -> Felino)
class Felino(Carnivoro):
    def __init__(self, nombre, edad, tipo_diente, patas):
        super().__init__(nombre, edad, tipo_diente)
        self.patas = patas

    def rugir(self):
        print(f"{self.nombre} ruge con un sonido imponente")

# Subclase: Canino (herencia multinivel: Animal -> Carn√≠voro -> Canino)
class Canino(Carnivoro):
    def __init__(self, nombre, edad, tipo_diente, cola):
        super().__init__(nombre, edad, tipo_diente)
        self.cola = cola

    def aullar(self):
        print(f"{self.nombre} a√∫lla al cielo")

# Subclase: Le√≥n (sobrescribe rugir)
class Leon(Felino):
    def __init__(self, nombre, edad, tipo_diente, patas, melena_color):
        super().__init__(nombre, edad, tipo_diente, patas)
        self.melena_color = melena_color

    def rugir(self):
        print("¬°El rey de la selva ruge con fuerza! ü¶Å")

# Subclase: Tigre (sobrescribe rugir)
class Tigre(Felino):
    def __init__(self, nombre, edad, tipo_diente, patas, rayas_color):
        super().__init__(nombre, edad, tipo_diente, patas)
        self.rayas_color = rayas_color

    def rugir(self):
        print("¬°El tigre hace temblar la selva con su rugido! üêØ")

# Subclase: Lobo (sobrescribe aullar)
class Lobo(Canino):
    def __init__(self, nombre, edad, tipo_diente, cola, pelaje_color):
        super().__init__(nombre, edad, tipo_diente, cola)
        self.pelaje_color = pelaje_color

    def aullar(self):
        print("¬°El lobo a√∫lla a la luna llena! üê∫")

# Subclase: Coyote (sobrescribe aullar)
class Coyote(Canino):
    def __init__(self, nombre, edad, tipo_diente, cola, pelaje_color):
        super().__init__(nombre, edad, tipo_diente, cola)
        self.pelaje_color = pelaje_color

    def aullar(self):
        print("¬°El coyote a√∫lla con astucia en la noche! üêæ")

# ===============================
# Herencia m√∫ltiple / h√≠brida
# ===============================

class Reptil(Animal):
    def arrastrarse(self):
        print(f"{self.nombre} se arrastra lentamente ü¶é")

class Volador(Animal):
    def volar(self):
        print(f"{self.nombre} est√° volando por los cielos üïäÔ∏è")

# Clase h√≠brida (hereda de Reptil y Volador)
class Dragon(Reptil, Volador):
    def lanzar_fuego(self):
        print(f"{self.nombre} lanza fuego üî•")

# ===============================
# Pruebas de la jerarqu√≠a
# ===============================

if __name__ == "__main__":
    simba = Leon("Simba", 5, "afilados", 4, "dorado")
    shere_khan = Tigre("Shere Khan", 8, "afilados", 4, "naranja con negro")
    akela = Lobo("Akela", 6, "afilados", True, "gris")
    willy = Coyote("Willy", 4, "afilados", True, "marr√≥n")
    draco = Dragon("Draco", 100)

    # Probar m√©todos heredados y sobrescritos
    simba.comer()
    simba.rugir()

    shere_khan.rugir()

    akela.aullar()
    willy.aullar()

    draco.arrastrarse()
    draco.volar()
    draco.lanzar_fuego()

    # Verificar MRO (Method Resolution Order) del drag√≥n
    print("\nMRO de Dragon:", [cls.__name__ for cls in Dragon.mro()])
```
